```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
v = np.array([1,2])
sk = [-0.5,1,1.5,3]
for i in sk:
    sk_v = i*v
    plt.plot([0, sk_v[0]], [0, sk_v[1]], 'o-', linewidth=2, label='skalar=%g' %i)
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
v1 = np.array([1,3])
v2 = np.array([3,1])

v3a = np.add(v1,v2)

print(v3a)
```

```python
plt.plot([0, v1[0]], [0, v1[1]], linewidth=2, label= 'v1')
plt.plot([0, v2[0]], [0, v2[1]], linewidth=2, label= 'v2')
plt.plot([0, v3a[0]], [0, v3a[1]], linewidth=2, label= 'v3a')
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([0, v1[0]], [0, v1[1]], linewidth=2, label= 'v1')
plt.plot([0, v2[0]] +v1[0], [0, v2[1]]+v1[1], linewidth=2, label= 'v2')
plt.plot([0, v3a[0]], [0, v3a[1]], linewidth=2, label= 'v3a')
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```
